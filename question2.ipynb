```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
train=pd.read_csv("train.csv")
train
```

```python
survived=train[train["Survived"]==1]
died=train[train["Survived"]==0]
```

```python
survived
```

```python
died
```

```python
plot.scatter(survived["Fare"],survived["Age"],label="Survivors",marker="v")
plot.scatter(died["Fare"],died["Age"],label="Casualties")
plot.xlabel("Fare")
plot.ylabel("Age")
plot.legend()
```

```python
# I have done the next part twice, once with a bar plot that has the ability to show colors, and once with a histogram
# which for the life of me, I can't figure out how to display colors. Since these are supposed to be histograms, I
# feel it is important to use the histogram function, however displaying colors would match the solution, therefore
# I displayed both options. Please let me know which one is correct. Thank you.
```

```python
survived_counts=survived["Embarked"].value_counts()
survived_counts=survived_counts.sort_index()
survived_counts
```

```python
plot.bar(survived_counts.index,survived_counts.values,width=.15,color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plot.ylabel("Survived")
plot.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
```

```python
died_counts=died["Embarked"].value_counts()
died_counts=died_counts.sort_index()
died_counts
```

```python
plot.bar(died_counts.index,died_counts.values,width=.15,color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plot.ylabel("Casualties")
plot.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
```
